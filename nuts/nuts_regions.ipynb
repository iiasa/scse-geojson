```python
import geopandas as gpd
```

```python
nuts_json = gpd.read_file('NUTS_RG_03M_2024_4326.geojson')
nuts_json.plot()
```

```python
nuts1_regions = nuts_json[nuts_json['LEVL_CODE'] == 1]

nuts1_regions.to_file('nuts1_regions.geojson', driver='GeoJSON')

nuts1_json = gpd.read_file('nuts1_regions.geojson')
nuts1_json.plot()
```

```python
nuts2_regions = nuts_json[nuts_json['LEVL_CODE'] == 2]

nuts2_regions.to_file('nuts2_regions.geojson', driver='GeoJSON')

nuts2_json = gpd.read_file('nuts2_regions.geojson')
nuts2_json.plot()
```

```python
nuts3_regions = nuts_json[nuts_json['LEVL_CODE'] == 3]

nuts3_regions.to_file('nuts3_regions.geojson', driver='GeoJSON')

nuts3_json = gpd.read_file('nuts3_regions.geojson')
nuts3_json.plot()
```
